```python
# Load satellite image and graph, then overlay nodes on the image
import pickle

# Load the satellite image (using region_1 to match the graph loaded above)
sat_img = cv2.imread(path + "region_1_sat.png")
sat_img = cv2.cvtColor(sat_img, cv2.COLOR_BGR2RGB)

# Load the graph (already loaded as 'thing' but let's reload to be clear)
graph = pickle.load(open(path + "region_1_graph_gt.pickle", "rb"))

# Create a REAL BIG figure
plt.figure(figsize=(24, 24))
plt.imshow(sat_img)

# Overlay all nodes as red dots
for node in graph.keys():
    y, x = node
    plt.plot(x, y, 'ro', markersize=3, alpha=0.7)

# Optionally overlay edges as green lines
for node, neighbors in graph.items():
    y1, x1 = node
    for neighbor in neighbors:
        y2, x2 = neighbor
        plt.plot([x1, x2], [y1, y2], 'g-', linewidth=0.5, alpha=0.3)

plt.title(f'Satellite Image with Graph Overlay ({len(graph)} nodes)', fontsize=20)
plt.axis('off')
plt.tight_layout()
plt.show()

print(f"Image shape: {sat_img.shape}")
print(f"Number of nodes: {len(graph)}")
```

```python
# Alternative: Just nodes overlay (cleaner, no edges)
plt.figure(figsize=(24, 24))
plt.imshow(sat_img)

# Overlay all nodes as red dots
for node in graph.keys():
    y, x = node
    plt.plot(x, y, 'ro', markersize=4, alpha=0.8)

plt.title(f'Satellite Image with Node Overlay ({len(graph)} junctions)', fontsize=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```
