```python
from OpenSERGE.openserge.data.dataset import CityScale 
from OpenSERGE.openserge.models.net import SingleShotRoadGraphNet
import torch
```

```python
net = SingleShotRoadGraphNet(backbone='resnet50')
net.load_state_dict(torch.load('OpenSERGE/checkpoints/junction_best.pt')['model_state_dict'])
```

```python
print(net)
```

```python
test_ds = CityScale(data_root='data', split='test')
sample = test_ds[0]
```

```python
sample
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(sample["image"].permute(1, 2, 0))
```

```python
out = net.forward(sample["image"].unsqueeze(0))
plt.imshow(out['junction_logits'].detach().numpy()[0,0,...])
```

```python
print(out['offset'].shape)

# Visualize offset map as quiver plot

plt.quiver(out['offset'].detach().numpy()[0,0,...], out['offset'].detach().numpy()[0,1,...])
```
